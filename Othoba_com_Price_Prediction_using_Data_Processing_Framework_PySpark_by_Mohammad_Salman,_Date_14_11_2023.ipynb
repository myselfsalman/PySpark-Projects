# ***Handling Missing values***

```python
from pyspark.sql.functions import mean
```

# ***Handling missing values for numerical columns: "Price', 'shippingPrice', 'Total', 'Ratings'" by mean***

# ***Calculate the mean of each numeric column***

```python
numeric_cols = ['Price', 'shippingPrice', 'Total', 'Ratings']


numeric_means = salman.select([mean(col).alias(col) for col in numeric_cols]).collect()[0]


for col in numeric_cols:
    salman = salman.na.fill(numeric_means[col], [col])
```

# ***Handling missing values for of each string column: 'Outlet_Size', 'Category', 'subCategory', 'Name', 'Brand', 'Seller', 'City' by Mode***

```python
from pyspark.sql.functions import when


string_cols = ['Outlet_Size', 'Category', 'subCategory', 'Name', 'Brand', 'Seller', 'City']


string_modes = [salman.groupBy(col).count().orderBy('count', ascending=False).first()[0] for col in string_cols]


for i, col in enumerate(string_cols):
    salman = salman.withColumn(col, when(salman[col].isNull(), string_modes[i]).otherwise(salman[col]))
```

# **Now, Finding Outliers values that may cause of lower accuracy on model Implementation for each feature sequentially from first to last**

# ***Using Z-score***

```python
from pyspark.sql.functions import *

column_subset = salman.columns

for col in column_subset:
    if salman.select(col).dtypes[0][1] == "string":
        pass
    else:
        col_mean = salman.select(mean(col)).collect()[0][0]
        col_stddev = salman.select(stddev(col)).collect()[0][0]
        upper_limit = col_mean + (3 * col_stddev)
        lower_limit = col_mean - (3 * col_stddev)

        outliers = salman.filter((salman[col] < lower_limit) | (salman[col] > upper_limit))

        if outliers.count() > 0:
            print(f"Outliers in column '{col}':")
            outliers.select(col).show()
```

# ***Dataframe values after filtering out the outliers using Z-score.***

```python
from pyspark.sql.functions import *

column_subset = salman.columns

for col in column_subset:
    if salman.select(col).dtypes[0][1] == "string":
        pass
    else:
        col_mean = salman.select(mean(col)).collect()[0][0]
        col_stddev = salman.select(stddev(col)).collect()[0][0]
        upper_limit = col_mean + (3 * col_stddev)
        lower_limit = col_mean - (3 * col_stddev)
        salman = salman.filter((salman[col] >= lower_limit) & (salman[col] <= upper_limit))

salman.show()
```

# ***01. Linear Regression***

```python
salman.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator


assembler = VectorAssembler(inputCols=["shippingPrice", "Total", "Ratings", "Day", "Month", "Year", "Category_index", "subCategory_index", "Name_index", "Brand_index", "Seller_index", "City_index", "Outlet_Size_index"], outputCol='features')
lr = LinearRegression(featuresCol='features', labelCol='Price')

pipeline = Pipeline(stages=[assembler, lr])
```

# ***Split the data into train and test***

```python
(train_data, test_data) = salman.randomSplit([0.8, 0.2], seed=123)
```

# ***Train the model***

```python
model = pipeline.fit(train_data)
```

# ***Making predictions***

```python
predictions = model.transform(test_data)
```

# ***Evaluate the model***

```python
evaluator = RegressionEvaluator(labelCol="Price", predictionCol="prediction", metricName="mse")
mse = evaluator.evaluate(predictions)
rmse = evaluator.evaluate(predictions, {evaluator.metricName: "rmse"})
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})

print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"Root Mean Squared Error (RMSE) on test data: {rmse}")
print(f"R-squared (R2) on test data: {r2}")
```

# ***Here, R-square value is: 0.9929***

# ***Scatter Plot of Predicted vs. Actual Values***

```python
import matplotlib.pyplot as plt
predictions_pd = predictions.select("Price", "prediction").toPandas()

plt.scatter(predictions_pd["Price"], predictions_pd["prediction"])
plt.title("Scatter Plot of Predicted vs. Actual Values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

plt.show()
```

# ***02. Gradient Boosting Regression***

```python
from pyspark.ml.regression import GBTRegressor
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler

feature_columns = ["shippingPrice", "Total", "Ratings", "Day", "Month", "Year", "Category_index", "subCategory_index", "Name_index", "Brand_index", "Seller_index", "City_index", "Outlet_Size_index"]

assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')
gbt = GBTRegressor(featuresCol='features', labelCol='Price', maxIter=10, maxBins=6659)

pipeline = Pipeline(stages=[assembler, gbt])

(train_data, test_data) = salman.randomSplit([0.8, 0.2], seed=123)

model = pipeline.fit(train_data)

predictions = model.transform(test_data)

evaluator = RegressionEvaluator(labelCol="Price", predictionCol="prediction", metricName="mse")
mse = evaluator.evaluate(predictions)
rmse = evaluator.evaluate(predictions, {evaluator.metricName: "rmse"})
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})

print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"Root Mean Squared Error (RMSE) on test data: {rmse}")
print(f"R-squared (R2) on test data: {r2}")
```

# ***Here, R-square value: 0.8323***

# ***03. Decision Trees Regression***

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator


feature_columns = ["shippingPrice", "Total", "Ratings", "Day", "Month", "Year", "Category_index", "subCategory_index", "Name_index", "Brand_index", "Seller_index", "City_index", "Outlet_Size_index"]

assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')

dt = DecisionTreeRegressor(featuresCol='features', labelCol='Price', maxBins=7000)


pipeline = Pipeline(stages=[assembler, dt])

model = pipeline.fit(train_data)
predictions = model.transform(test_data)

evaluator = RegressionEvaluator(labelCol='Price', predictionCol='prediction', metricName='mse')
mse = evaluator.evaluate(predictions)

rmse = evaluator.evaluate(predictions, {evaluator.metricName: 'rmse'})

r2 = evaluator.evaluate(predictions, {evaluator.metricName: 'r2'})

print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"Root Mean Squared Error (RMSE) on test data: {rmse}")
print(f"R-squared (R2) on test data: {r2}")
```

# ***Here, R-square value: 0.6960***

# ***04. Generalized Linear Regression (GLR):***

```python
from pyspark.sql import SparkSession
from pyspark.ml.regression import GeneralizedLinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.feature import VectorAssembler

feature_columns = [col_name for col_name in salman.columns if col_name != 'Price']
assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')
salman = assembler.transform(salman)

(trainingData, testData) = salman.randomSplit([0.8, 0.2], seed=42)

glr = GeneralizedLinearRegression(featuresCol='features', labelCol='Price', family='gaussian', link='identity')

model = glr.fit(trainingData)

predictions = model.transform(testData)

evaluator = RegressionEvaluator(labelCol='Price', predictionCol='prediction', metricName='rmse')
rmse = evaluator.evaluate(predictions)
mse = evaluator.evaluate(predictions, {evaluator.metricName: "mse"})
r2 = evaluator.evaluate(predictions, {evaluator.metricName: "r2"})


print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"Root Mean Squared Error (RMSE) on test data: {rmse}")
print(f"R-squared (R2) on test data: {r2}")
```

# ***Here, R-square value: 0.9966***

# ***05. Lasso Regression***

```python
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import RegressionEvaluator


feature_columns = [col_name for col_name in salman.columns if col_name != 'Price']
assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')

lasso = LinearRegression(featuresCol='features', labelCol='Price', elasticNetParam=1.0, regParam=0.01)

pipeline_lasso = Pipeline(stages=[assembler, lasso])

(train_data, test_data) = salman.randomSplit([0.8, 0.2], seed=123)

model_lasso = pipeline_lasso.fit(train_data)

predictions_lasso = model_lasso.transform(test_data)

evaluator_lasso = RegressionEvaluator(labelCol="Price", predictionCol="prediction", metricName="rmse")
rmse_lasso = evaluator_lasso.evaluate(predictions_lasso)
mse = evaluator_lasso.evaluate(predictions_lasso, {evaluator_lasso.metricName: "mse"})
r2_lasso = evaluator_lasso.evaluate(predictions_lasso, {evaluator_lasso.metricName: "r2"})

print(f"Root Mean Squared Error: {rmse_lasso}")
print(f"Mean Squared Error (MSE) on test data: {mse}")
print(f"R-squared (R2): {r2_lasso}")
```
